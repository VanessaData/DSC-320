#### A Riemann sum is an estimation of the area under a curve using rectangles. The right Riemann sum Rn of a function f on the interval [a, b] with n rectangles is calculated by
#### Xn Rn = f(a + Δx)Δx + f(a + 2Δx)Δx + · · · + f(a + nΔx)Δx = f(a + kΔx)Δx,k=1
#### where Δx = b−a , is the width of the rectangles. Write a Python function to take in a mathematical function f, two interval endpoints a and b, and the number of rectangles n, and then outputs the right Riemann sum
#### Rn

```python
import numpy as np
```

```python
# Define the right Riemann sum function
def right_riemann_sum(f, a, b, n):
    # Calculate the width of each rectangle
    dx = (b - a) / n
    
    # Generate right endpoints
    x_values = np.linspace(a + dx, b, n)
    
    # Calculate the sum of the areas of the rectangles
    Rn = np.sum(f(x_values) * dx)
    
    return Rn

# Example usage
# Define the function f(x)
def f(x):
    return 4 - x**2

# Interval endpoints
a = 0
b = 2

# Number of rectangles
n = 10

# Calculate the right Riemann sum
Rn = right_riemann_sum(f, a, b, n)
print(f"The right Riemann sum with {n} rectangles is: {Rn:.6f}")
```
