### Section 1

#### Part A

```python
import pandas as pd

# Load the data
file_path = 'qb_stats.csv'
qb_data = pd.read_csv(file_path)

# Calculate the mean of each numerical column
mean_values = qb_data.mean(numeric_only=True)
print(mean_values)
```

#### Part B

```python
# Calculate the standard deviation of each numerical column
std_values = qb_data.std(numeric_only=True)
print(std_values)
```

#### Part C

```python
import matplotlib.pyplot as plt

# Create a histogram for the 'yds' (yards) column
plt.figure(figsize=(10, 6))
plt.hist(qb_data['yds'], bins=15, edgecolor='black')
plt.title("Histogram of Yards (yds)")
plt.xlabel("Yards")
plt.ylabel("Frequency")
plt.show()
```

#### Part D

```python
# Create a boxplot for the 'td' (touchdowns) column
plt.figure(figsize=(8, 6))
plt.boxplot(qb_data['td'], vert=False)
plt.title("Boxplot of Touchdowns (td)")
plt.xlabel("Touchdowns")

# Display the plot
plt.show()

# Identifying outliers
q1 = qb_data['td'].quantile(0.25)
q3 = qb_data['td'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Finding outliers
outliers = qb_data[(qb_data['td'] < lower_bound) | (qb_data['td'] > upper_bound)]['td']
print(outliers)
```

### Section 2

#### Part A

```python
# Load the survey data
survey_data = pd.read_csv('survey_data.csv')

# Calculate the probability of a college student having brown hair
total_students = survey_data.shape[0]
brown_hair_count = survey_data[survey_data['hair_color'].str.lower() == 'brown'].shape[0]
brown_hair_probability = brown_hair_count / total_students

print(brown_hair_probability)
```

#### Part B

```python
# Calculate the probability of a college student having blue eyes
blue_eyes_count = survey_data[survey_data['eye_color'].str.lower() == 'blue'].shape[0]
blue_eyes_probability = blue_eyes_count / total_students

print(blue_eyes_probability)
```

#### Part C

```python
# Calculate the probability of a college student having blue eyes given that they have brown hair
brown_hair_blue_eyes_count = survey_data[
    (survey_data['hair_color'].str.lower() == 'brown') &
    (survey_data['eye_color'].str.lower() == 'blue')
].shape[0]

# Total students with brown hair
brown_hair_count = survey_data[survey_data['hair_color'].str.lower() == 'brown'].shape[0]

# Conditional probability: P(Blue Eyes | Brown Hair)
blue_eyes_given_brown_hair_probability = brown_hair_blue_eyes_count / brown_hair_count

print(blue_eyes_given_brown_hair_probability)
```

### Part D

```python
# Calculate the probability of a college student having brown hair given that they have blue eyes
blue_eyes_brown_hair_count = survey_data[
    (survey_data['eye_color'].str.lower() == 'blue') &
    (survey_data['hair_color'].str.lower() == 'brown')
].shape[0]

# Total students with blue eyes
blue_eyes_count = survey_data[survey_data['eye_color'].str.lower() == 'blue'].shape[0]

# Conditional probability: P(Brown Hair | Blue Eyes)
brown_hair_given_blue_eyes_probability = blue_eyes_brown_hair_count / blue_eyes_count

print(brown_hair_given_blue_eyes_probability)
```

### Part E

#### Since these values differ, the events "having brown hair" and "having blue eyes" are not independent. This discrepancy indicates that the occurrence of one event does affect the likelihood of the other in this dataset.

#### Part F

```python
# Count values for each category in 'hair_color' and 'eye_color'
hair_color_counts = survey_data['hair_color'].value_counts()
eye_color_counts = survey_data['eye_color'].value_counts()

# Plot the hair color distribution
plt.figure(figsize=(10, 6))
hair_color_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Distribution of Hair Colors")
plt.xlabel("Hair Color")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.show()

# Plot the eye color distribution
plt.figure(figsize=(10, 6))
eye_color_counts.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title("Distribution of Eye Colors")
plt.xlabel("Eye Color")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.show()
```
